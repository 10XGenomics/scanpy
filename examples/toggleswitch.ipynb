# Simulate and Analyze Toggleswitch

A simple toggle switch of mutually suppressing genes, as in Nature 403, 339-342 (2000).

```python
from sys import path
path.insert(0,'..')
import scanpy as sc

# set very low png resolution, to decrease storage space
sc.sett.dpi(30)
# show some output
sc.sett.verbosity = 1
```

```python
# let us first simulate an extremely simplet model, 
# a 'toggleswitch'
params = sc.read_params('../sim/toggleswitch_params.txt')
# pass params as keyword arguments
ddata = sc.sim(**params)
sc.plot(ddata)
```

```python
# define root cell for pseudotime analysis
# simply take the first data point (the first row of the
# data matrix X) for that
ddata['xroot'] = ddata['X'][0]
ddpt = sc.dpt(ddata)
sc.plot(ddpt, ddata)
```
